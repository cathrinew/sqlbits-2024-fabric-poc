```sql
create or replace temporary view listing_transform as
SELECT 
l.Id AS listingId, 
name, 
Host_id as hostid,
n.id as neighbourhoodid,
latitude, 
longitude, 
r.id as roomtypeid,
price, 
minimum_nights as minimumnights, 
number_of_reviews AS numberofreviews,
last_review AS lastreview, 
CASE WHEN Reviews_per_month is NULL THEN -1 ELSE Reviews_per_month END AS reviewspermonth, 
Calculated_host_listings_count AS calculatedhostlistingscount, 
availability_365 AS availability365, 
number_of_reviews_ltm AS numberofreviewsltm, 
CASE WHEN license is NULL THEN -1 ELSE license END AS licens

FROM Bronze_AirBnB_Lakehouse.listings l

left join Silver_airbnb_lakehouse.dimneighbourhood as n on l.neighbourhood = n.neighbourhood
left join Silver_airbnb_lakehouse.dimroomtypes as r on l.room_type = r.roomtype
```

```sql
MERGE INTO Silver_AirBnB_Lakehouse.dimlistings USING listing_transform ON Silver_AirBnB_Lakehouse.dimlistings.listingId = listing_transform.listingId
WHEN MATCHED THEN UPDATE SET *
WHEN NOT MATCHED THEN INSERT *;
```
