```sql
CREATE OR REPLACE TEMPORARY VIEW RoomTypeCategory AS
SELECT 
Id,
RoomType,
CASE WHeN lower(RoomType) in ('private room', 'shared room') THEN 'Cheap' 
     WHEN lower(RoomType) in ('entire home/apt', 'hotel room') THEN 'Expensive' END AS RoomTypeCategory
FROM Silver_AirBnB_Lakehouse.dimroomtypes
```

```sql
MERGE INTO Gold_AirBnB_Lakehouse.dimroomtypes USING RoomTypeCategory ON Gold_AirBnB_Lakehouse.dimroomtypes.Id = RoomTypeCategory.Id
WHEN MATCHED THEN UPDATE SET *
WHEN NOT MATCHED THEN INSERT *;
```
