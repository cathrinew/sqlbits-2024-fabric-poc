```sql
CREATE OR REPLACE TEMPORARY VIEW Fancy_Neighbourhood AS
SELECT 
Id, 
Neighbourhood, 
CASE WHEN lower(Neighbourhood) = 'kensington and chelsea' or lower(Neighbourhood) = 'city of london' or lower (Neighbourhood) = 'camden' THEN 'Fancy' 
     ELSE 'Standard' END AS NeighbourhoodCategory
FROM Silver_AirBnB_Lakehouse.dimneighbourhood
```

```sql
MERGE INTO Gold_AirBnB_Lakehouse.dimneighbourhood USING Fancy_Neighbourhood ON Gold_AirBnB_Lakehouse.dimneighbourhood.Id = Fancy_Neighbourhood.Id
WHEN MATCHED THEN UPDATE SET *
WHEN NOT MATCHED THEN INSERT *;
```
