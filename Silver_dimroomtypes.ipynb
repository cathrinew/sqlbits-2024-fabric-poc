```sql
Create or replace temporary view RoomTypes_transform AS 
select 
row_number() OVER(ORDER BY Room_type) AS Id,
room_type AS RoomType
from (SELECT distinct room_type from Bronze_AirBnB_Lakehouse.listings)
```

```sql
MERGE INTO Silver_AirBnB_Lakehouse.dimroomtypes USING RoomTypes_transform ON Silver_AirBnB_Lakehouse.dimroomtypes.Id = RoomTypes_transform.Id 
WHEN MATCHED THEN UPDATE SET *
WHEN NOT MATCHED THEN INSERT *;
```
