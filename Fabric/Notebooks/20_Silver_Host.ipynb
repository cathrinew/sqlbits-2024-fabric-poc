```sql
CREATE OR REPLACE TEMPORARY VIEW vHost AS
SELECT DISTINCT
    host_id AS Id,
    host_name AS Name,
    CASE host_is_superhost WHEN 't' THEN 'Superhost' ELSE 'Host' END AS HostType,
    host_since AS HostSince,
    host_neighbourhood AS Neighbourhood,
    CASE INSTR(host_location, ',') WHEN 0 THEN NULL ELSE TRIM(LEFT(host_location, INSTR(host_location, ',') - 1)) END AS City,
    CASE INSTR(host_location, ',') WHEN 0 THEN host_location ELSE TRIM(RIGHT(host_location, LEN(host_location) - INSTR(host_location, ','))) END AS Country,
    CASE host_response_time WHEN 'N/A' THEN NULL ELSE host_response_time END AS ResponseTime,
    CASE host_response_rate WHEN 'N/A' THEN NULL ELSE (REPLACE(host_response_rate, '%', '') / 100.00) END AS ResponseRate,
    CASE host_acceptance_rate WHEN 'N/A' THEN NULL ELSE (REPLACE(host_acceptance_rate, '%', '') / 100.00) END AS AcceptanceRate,
    host_url AS HostURL,
    host_picture_url AS HostPictureURL,
    host_thumbnail_url AS HostPictureThumbnailURL,
    host_about AS About
FROM Bronze_Airbnb.listings_details
```

```sql
MERGE INTO Silver_Airbnb.Host USING vHost ON Silver_Airbnb.Host.Id = vHost.Id
WHEN MATCHED THEN UPDATE SET *
WHEN NOT MATCHED THEN INSERT *
```
