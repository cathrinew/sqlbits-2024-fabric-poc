```sql
CREATE OR REPLACE TEMPORARY VIEW vNeighbourhoods AS 
SELECT DISTINCT
    DENSE_RANK() OVER (ORDER BY neighbourhood) AS Id,
    neighbourhood AS Neighbourhood
FROM Bronze_Airbnb.neighbourhoods
```

```sql
MERGE INTO Silver_Airbnb.Neighbourhood USING vNeighbourhoods ON Silver_Airbnb.Neighbourhood.Id = vNeighbourhoods.Id
WHEN MATCHED THEN UPDATE SET *
WHEN NOT MATCHED THEN INSERT *
```
