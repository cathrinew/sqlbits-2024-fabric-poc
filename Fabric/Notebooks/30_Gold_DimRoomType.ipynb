```sql
CREATE OR REPLACE TEMPORARY VIEW vDimRoomType AS
SELECT 
     Id AS RoomTypeKey,
     RoomType,
     CASE
          WHEN lower(RoomType) in ('private room', 'shared room') THEN 'Cheap' 
          WHEN lower(RoomType) in ('entire home/apt', 'hotel room') THEN 'Expensive ðŸ’°'
     END AS RoomTypeCategory
FROM Silver_Airbnb.RoomType
```

```sql
MERGE INTO Gold_Airbnb.DimRoomType USING vDimRoomType ON Gold_Airbnb.DimRoomType.RoomTypeKey = vDimRoomType.RoomTypeKey
WHEN MATCHED THEN UPDATE SET *
WHEN NOT MATCHED THEN INSERT *
```
