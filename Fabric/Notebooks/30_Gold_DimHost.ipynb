```sql
CREATE OR REPLACE TEMPORARY VIEW vDimHost AS
SELECT 
     Id AS HostKey,
     Name AS HostName,
     HostType,
     HostSince,
     ResponseRate AS HostResponseRate,
     AcceptanceRate AS HostAcceptanceRate,
     HostURL 
FROM Silver_Airbnb.Host
```

```sql
MERGE INTO Gold_Airbnb.DimHost USING vDimHost ON Gold_Airbnb.DimHost.HostKey = vDimHost.HostKey
WHEN MATCHED THEN UPDATE SET *
WHEN NOT MATCHED THEN INSERT *
```
