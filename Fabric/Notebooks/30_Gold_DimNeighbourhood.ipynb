```sql
CREATE OR REPLACE TEMPORARY VIEW vDimNeighbourhood AS
SELECT 
     Id AS NeighbourhoodKey, 
     Neighbourhood, 
     CASE 
          WHEN lower(Neighbourhood) = 'kensington and chelsea' OR lower(Neighbourhood) = 'city of london' OR lower (Neighbourhood) = 'camden' THEN 'Fancy âœ¨' 
          ELSE 'Standard' 
     END AS NeighbourhoodCategory
FROM Silver_Airbnb.Neighbourhood
```

```sql
MERGE INTO Gold_Airbnb.DimNeighbourhood USING vDimNeighbourhood ON Gold_Airbnb.DimNeighbourhood.NeighbourhoodKey = vDimNeighbourhood.NeighbourhoodKey
WHEN MATCHED THEN UPDATE SET *
WHEN NOT MATCHED THEN INSERT *;
```
