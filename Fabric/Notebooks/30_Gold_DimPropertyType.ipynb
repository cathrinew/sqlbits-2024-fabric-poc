```sql
CREATE OR REPLACE TEMPORARY VIEW vDimPropertyType AS
SELECT
    Id AS PropertyTypeKey,
    PropertyType,
    CASE
        WHEN 
            LOWER(PropertyType) LIKE '%hostel%' 
            OR LOWER(PropertyType) LIKE '%shared room%' 
            THEN 'Budget'
        WHEN
            LOWER(PropertyType) LIKE '%castle%' 
            OR LOWER(PropertyType) LIKE '%entire%' 
            OR LOWER(PropertyType) LIKE '%suite%' 
            THEN 'Exclusive ðŸŒŸ'
        WHEN 
            LOWER(PropertyType) LIKE '%barn%'
            OR LOWER(PropertyType) LIKE '%boat%'
            OR LOWER(PropertyType) LIKE '%camp%'
            OR LOWER(PropertyType) LIKE '%cave%'
            OR LOWER(PropertyType) LIKE '%container%'
            OR LOWER(PropertyType) LIKE '%dome%'
            OR LOWER(PropertyType) LIKE '%earth%'
            OR LOWER(PropertyType) LIKE '%farm%'
            OR LOWER(PropertyType) LIKE '%island%'
            OR LOWER(PropertyType) LIKE '%lighthouse%'
            OR LOWER(PropertyType) LIKE '%religious%'
            OR LOWER(PropertyType) LIKE '%riad%'
            OR LOWER(PropertyType) LIKE '%shepherd%'
            OR LOWER(PropertyType) LIKE '%tent%'
            OR LOWER(PropertyType) LIKE '%tiny%'
            OR LOWER(PropertyType) LIKE '%tower%'
            OR LOWER(PropertyType) LIKE '%tree%'
            OR LOWER(PropertyType) LIKE '%yurt%'
            THEN 'Unconventional'
        ELSE 'Standard'
    END AS PropertyTypeCategory
FROM Silver_Airbnb.PropertyType
```

```sql
MERGE INTO Gold_Airbnb.DimPropertyType USING vDimPropertyType ON Gold_Airbnb.DimPropertyType.PropertyTypeKey = vDimPropertyType.PropertyTypeKey
WHEN MATCHED THEN UPDATE SET *
WHEN NOT MATCHED THEN INSERT *
```
