```sql
CREATE OR REPLACE TEMPORARY VIEW vRoomType AS 
SELECT DISTINCT
    DENSE_RANK() OVER (ORDER BY room_type) AS Id,
    room_type AS RoomType
FROM  Bronze_Airbnb.listings_details
```

```sql
MERGE INTO Silver_Airbnb.RoomType USING vRoomType ON Silver_Airbnb.RoomType.Id = vRoomType.Id 
WHEN MATCHED THEN UPDATE SET *
WHEN NOT MATCHED THEN INSERT *
```
