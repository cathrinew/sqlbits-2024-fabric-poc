```sql
CREATE OR REPLACE TEMPORARY VIEW vListings AS
SELECT
    id AS Id,
    name AS Name,
    host_id as HostId,
    host_since as HostSince,
    neighbourhood_cleansed AS Neighbourhood,
    DENSE_RANK() OVER (ORDER BY neighbourhood_cleansed) as NeighbourhoodId,
    latitude AS Latitude,
    longitude AS Longitude,
    room_type AS RoomType,
    DENSE_RANK() OVER (ORDER BY room_type) as RoomTypeId,
    property_type AS PropertyType,
    DENSE_RANK() OVER (ORDER BY property_type) as PropertyTypeId,
    instant_bookable AS InstantBookable,
    listing_url AS URL,
    CAST(REPLACE(price, '$', '') AS DECIMAL(14,2)) AS Price,
    minimum_nights as MinimumNights,
    number_of_reviews AS NumberOfReviews,
    last_review AS LastReview,
    CASE WHEN reviews_per_month IS NULL THEN -1 ELSE reviews_per_month END AS ReviewsPerMonth,
    calculated_host_listings_count AS CalculatedHostListingsCount,
    availability_365 AS Availability365,
    number_of_reviews_ltm AS NumberOfReviewsLtm,
    CASE WHEN license IS NULL THEN -1 ELSE license END AS License
FROM Bronze_Airbnb.listings_details
```

```sql
MERGE INTO Silver_Airbnb.Listings USING vListings ON Silver_Airbnb.Listings.Id = vListings.Id
WHEN MATCHED THEN UPDATE SET *
WHEN NOT MATCHED THEN INSERT *
```
