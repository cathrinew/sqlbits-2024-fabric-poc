```sql
CREATE OR REPLACE TEMPORARY VIEW vFactListings AS 
SELECT 
    DATE_FORMAT(HostSince, 'yyyyMMdd') AS DateKey,
    HostId AS HostKey,
    Id AS ListingKey, 
    NeighbourhoodId AS NeighbourhoodKey,
    PropertyTypeId AS PropertyTypeKey, 
    RoomTypeId AS RoomTypeKey, 
    Price, 
    MinimumNights, 
    NumberOfReviews, 
    LastReview, 
    ReviewsPerMonth, 
    CalculatedHostListingsCount, 
    Availability365
FROM Silver_Airbnb.Listings
```

```sql
MERGE INTO Gold_Airbnb.FactListings USING vFactListings on Gold_Airbnb.FactListings.ListingKey = vFactListings.ListingKey
WHEN MATCHED THEN UPDATE SET *
WHEN NOT MATCHED THEN INSERT *;
```
