```sql
CREATE OR REPLACE TEMPORARY VIEW vDimListing AS 
SELECT 
    Id AS ListingKey, 
    Name AS ListingName, 
    URL AS ListingURL,
    Latitude AS ListingLatitude, 
    Longitude AS ListingLongitude, 
    InstantBookable AS IsListingInstantBookable
FROM Silver_Airbnb.Listings
```

```sql
MERGE INTO Gold_Airbnb.DimListing USING vDimListing on Gold_Airbnb.DimListing.ListingKey = vDimListing.ListingKey
WHEN MATCHED THEN UPDATE SET *
WHEN NOT MATCHED THEN INSERT *
```
