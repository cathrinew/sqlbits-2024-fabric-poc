```sql
CREATE OR REPLACE TEMPORARY VIEW vPropertyType AS 
SELECT DISTINCT
    DENSE_RANK() OVER (ORDER BY property_type) AS Id,
    property_type AS PropertyType
FROM Bronze_Airbnb.listings_details
```

```sql
MERGE INTO Silver_Airbnb.PropertyType USING vPropertyType ON Silver_Airbnb.PropertyType.Id = vPropertyType.Id 
WHEN MATCHED THEN UPDATE SET *
WHEN NOT MATCHED THEN INSERT *
```
