```sql
Create or replace temporary view Neighbourhoods_transform AS 
select 
row_number() OVER(ORDER BY Neighbourhood) AS Id,
Neighbourhood
from (SELECT distinct Neighbourhood from Bronze_AirBnB_Lakehouse.neighbourhoods)
```

```sql
MERGE INTO Silver_AirBnB_Lakehouse.dimneighbourhood USING Neighbourhoods_transform ON Silver_AirBnB_Lakehouse.dimneighbourhood.Id = Neighbourhoods_transform.Id
WHEN MATCHED THEN UPDATE SET *
WHEN NOT MATCHED THEN INSERT *;
```
