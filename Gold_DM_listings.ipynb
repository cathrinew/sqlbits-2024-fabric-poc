```sql
CREATE OR REPLACE TEMPORARY VIEW dimlisting AS 
SELECT 
Listingid, 
Name, 
Latitude, 
Longitude, 
Licens
FROM Silver_AirBnB_Lakehouse.dimlistings
```

```sql
MERGE INTO Gold_AirBnB_Lakehouse.dimlistings USING dimlisting on Gold_AirBnB_Lakehouse.dimlistings.ListingId = dimlisting.ListingId
WHEN MATCHED THEN UPDATE SET *
WHEN NOT MATCHED THEN INSERT *;
```

```sql
CREATE OR REPLACE TEMPORARY VIEW factlisting AS 
SELECT 
Listingid, 
date_format(LastReview, 'yyyyMMdd') AS DateKey,
HostId, 
NeighbourhoodId, 
RoomTypeId, 
price, 
minimumNights, 
NumberOfReviews, 
LastReview, 
ReviewsPerMonth, 
CalculatedHostListingsCount, 
availability365
FROM Silver_AirBnB_Lakehouse.dimlistings
```

```sql
MERGE INTO Gold_AirBnB_Lakehouse.factlistings USING factlisting on Gold_AirBnB_Lakehouse.factlistings.ListingId = factlisting.ListingId
WHEN MATCHED THEN UPDATE SET *
WHEN NOT MATCHED THEN INSERT *;
```
